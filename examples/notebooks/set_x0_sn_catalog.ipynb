```python
import pandas as pd
from lsst.sims.photUtils import Bandpass
import sncosmo
from astropy.cosmology import FlatLambdaCDM
```

```python
dc2_df = pd.read_csv('/Users/Bryce/Documents/DESC/Twinkles/data/dc2_sne_cat.csv', index_col=0)
```

```python
dc2_df['MB']
```

```python
#DC2 Cosmology
#FlatLambdaCDM(H0=71 km / (Mpc s), Om0=0.265, Tcmb0=0 K, Neff=3.04, m_nu=None, Ob0=0.0448)
```

```python
source = sncosmo.get_source('salt2-extended')
```

```python
model = sncosmo.Model(source=source)
```

```python
model.param_names
```

```python
model.parameters
```

```python
# Danny's x1
model.set(x1 = 1.)
```

```python
cosmo = FlatLambdaCDM(H0=71, Om0=0.265, Tcmb0=0, Neff=3.04, m_nu=None, Ob0=0.0448)
```

```python
cosmo
```

```python
x0 = []
for sn_row in dc2_df.iterrows():
    z = sn_row[1]['zs']
    MB = sn_row[1]['MB']
    model.set(z=z)
    model.set_source_peakabsmag(MB, 'bessellb', 'ab', cosmo=cosmo)
    x0.append(model.get('x0'))
```

```python
x0
```

```python
dc2_df['x0'] = x0
```

```python
dc2_df['x1'] = 1.
dc2_df['c'] = 0.
```

```python
dc2_df.to_csv('/Users/Bryce/Documents/DESC/Twinkles/data/dc2_sne_cat.csv', index=False)
```
