```python
# Create a series of postage stamps from a list of 5 Exposure
# FITS files (using the test image as a proxy for all five) 
# and display each.
plt.rcParams['figure.figsize'] = (5, 5)
dataFolder = '../../data/Coadds/'#'/Users/Bryce/Documents/DESC/Twinkles/data/Coadds/'
expfiles=[]
bandpassFiles = {}
filterNames = []
for folderName in os.listdir(dataFolder):
    for fileName in os.listdir(str(dataFolder+folderName)):
        expfiles.append(str(dataFolder+folderName+'/'+fileName))
        bandpassFiles[folderName] = str(dataFolder+folderName+'/'+fileName)
        filterNames.append(folderName)
#loc0 = [52.9230946, -27.4559684]
#loc0 = [52.9252737, -27.5212009]
loc0 = [52.9968499, -27.4517142]
ra = [loc0[0]]
dec = [loc0[1]]
size = 20
rgb_dict = {}
for ra_i, dec_i in zip(ra,dec):
    stamps = create_postage_stamps(ra_i, dec_i, size, expfiles)
    for i, stamp in enumerate(stamps):
        outfile = 'stamp_%s.fits' % filterNames[i]
        stamp.writeFits(outfile)
        fits_obj = fits.open(outfile)
        fig = render_fits_image(fits_obj[1], title=outfile)
        rgb_dict[filterNames[i]] = fits_obj[1].data
        plt.savefig('stamp_%i.png' % (i))
    rgb_data = np.zeros((np.shape(rgb_dict['i'])[0], np.shape(rgb_dict['i'])[1], 3))
    for j,filterName in enumerate(['i', 'g', 'r']):
        rgb_data[:,:,j] = rgb_dict[filterName]/18#/111.99#np.max(rgb_dict[filterName])
    #rgb_data = np.array((rgb_dict['i'], rgb_dict['r'], rgb_dict['g']))
print np.shape(rgb_data)
#plt.imshow(rgb_data, origin='lower')#, norm=mpl.colors.LogNorm(), interpolation=None)
    #img.imsave('stamp_rgb.png', rgb_data.T)
```

```python
for fitsFile in expfiles:
    g=0
    r=0
    i=0
    hdulist = fits.open(fitsFile)
    if fitsFile[-6] == 'g':
        if g == 0:
            g_prime = hdulist[0]
            g_header = hdulist[1].header
            g_data = hdulist[1].data#/np.nanmax(hdulist[1].data)
            g_rest = hdulist[2:]
        else:
            g_data += hdulist[1].data#/np.nanmax(hdulist[1].data)
        g+=1
    elif fitsFile[-6] == 'r':
        if r == 0:
            r_prime = hdulist[0]
            r_header = hdulist[1].header
            r_data = hdulist[1].data#/np.nanmax(hdulist[1].data)
            r_rest = hdulist[2:]
        else:
            r_data = hdulist[1].data#/np.nanmax(hdulist[1].data)
        r+=1
    elif fitsFile[-6] == 'i':
        if i == 0:
            i_prime = hdulist[0]
            i_header = hdulist[1].header
            i_data = hdulist[1].data#/np.nanmax(hdulist[1].data)
            i_rest = hdulist[2:]
        else:
            i_data = hdulist[1].data#/np.nanmax(hdulist[1].data)
        i+=1
gPrimaryHDU = fits.ImageHDU(g_data, header=g_header)
rPrimaryHDU = fits.ImageHDU(r_data, header=r_header)
iPrimaryHDU = fits.ImageHDU(i_data, header=i_header)
gHDU = fits.HDUList([g_prime, gPrimaryHDU] + g_rest)
rHDU = fits.HDUList([r_prime, rPrimaryHDU] + r_rest)
iHDU = fits.HDUList([i_prime, iPrimaryHDU] + i_rest)
os.remove('g.fits')
os.remove('r.fits')
os.remove('i.fits')
gHDU.writeto('g.fits')
rHDU.writeto('r.fits')
iHDU.writeto('i.fits')
#newPrimaryHDU = fits.ImageHDU(hdulist[1].data + hdulist2[1].data, header=hdulist[1].header)
#hdulist3 = fits.HDUList([hdulist[0], newPrimaryHDU] + hdulist[2:])
#import shutil
#try:
#    shutil.rmtree('new.fits')
#except:
#    print 'here'
#hdulist3.writeto('new.fits')
```
